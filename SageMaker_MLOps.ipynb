## Build Serverless Endpoint by SageMaker
### requirments
* packages to be installed by pip
  + huggingface_hub==0.1.0
  + transformers==4.12
  + boto3
  + awscli
* set up aws config with access key and secret access key
* import models and pipeline from transformers
